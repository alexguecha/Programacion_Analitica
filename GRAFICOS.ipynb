### ***SUBPLOT***

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'m--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)       # 1 fila, dos columnas y se esta referenciando el número 1
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')       # Con la referencia aún al plot 1, se dibuja otra gŕafica
plt.subplot(1,2,2)       # ahora se referencia el plot número 2
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```

```python
fig = plt.figure()   # Creamos una figura (canvas vacío) 
ejes = fig.add_axes([0.1,0.1,0.5,0.9])   #  Añadimos los ejes
ejes.plot(x,y,'b')     # 'b' blue
fig.show()
```

```python
fig = plt.figure()

eje = fig.add_axes([0.1,0.1,0.8,0.9]) # eje principal
eje2 = fig.add_axes([0.2,0.55,0.4,0.3]) # eje interior

# Figura principal 
eje.plot(x,y,'b')

# Inserta figura 
eje2.plot(y,x,'r')
eje2.set_facecolor('gray')

fig.show()
```

## **SUBPLOTS**
Orientado a objetos

```python
# Se usa de forma similar a plt.figure() excepto que devuelve al mismo tiempo un objeto figura y un objeto (o un arreglo) de ejes
fig, eje = plt.subplots()
eje.plot(x,y,'b')

```

```python
# Ahora se usa el eje, como antes, para graficar en él
```

```python
fig, eje = plt.subplots(nrows=1, ncols=2)
eje[0].plot(x,y,'b')
eje[1].plot(y,x,'r')
```

```python
fig, (x1,x2) = plt.subplots(nrows=1, ncols=2)
x1.plot(x,y,'b')
x2.plot(y,x,'r')
```

```python
fig, ejes = plt.subplots(2, 4)
ejes[0,0].plot(x,np.cos(x))
ejes[0,1].plot(x,np.sin(x),'y')
ejes[0,2].plot(x,np.tan(x),'g')
ejes[0,3].plot(x,np.cos(x)**2)
```

```python
fig, ((x1,x2,x3,x4),(x5,x6,x7,x8)) = plt.subplots(2, 4)
x1.plot(x,np.cos(x))
x2.plot(x,np.sin(x),'y')
x3.plot(x,np.tan(x),'r')
x4.plot(x,np.cos(x)**2)
fig.tight_layout()    # Para evitar superposición de elementos
```

### ***Personalizando nuestros gráficos***

```python
fig, axes = plt.subplots(1,2,figsize=(5,5))
axes[0].plot(x,y,label="$(x)$")
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x,y,label='x2')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right', bbox_to_anchor=(0.2,0.2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")

tips.head()
```

STAT:
    count: show the number of observations in each bin

    frequency: show the number of observations divided by the bin width

    probability: or proportion: normalize such that bar heights sum to 1

    percent: normalize such that bar heights sum to 100

    density: normalize such that the total area of the histogram equals 1

```python
sns.histplot(data=tips, x='tip',bins=15,cumulative=False,hue='sex',stat='percent',multiple='dodge')
# multiple=Stack, layer
plt.show()
```

bw_adjust -- smooth

```python
sns.kdeplot(data=tips,x='tip',hue='sex',cumulative=False,shade=True,bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='hist',multiple='stack') #kde
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
plt.figure(figsize=(8,8))
markers = {"Lunch":"D", "Dinner":"s"}
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',col='time')
plt.show()
```
